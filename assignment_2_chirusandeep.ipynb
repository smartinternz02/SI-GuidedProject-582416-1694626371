```python
import seaborn as sns
import matplotlib.pyplot as plt

car_crashes = sns.load_dataset("car_crashes")
car_crashes.head()
```

```python
car_crashes.tail()
```

```python
# Correlation Heatmap
corr_matrix = car_crashes.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

**Inference for Correlation Heatmap:**

The above graph is Heatmap. it shows the correlation between numerical variables. For example, "alcohol" and "total" have a strong positive correlation.


```python
# Bar Plot
sns.barplot(x="total", y="abbrev", data=car_crashes.sort_values(by="total", ascending=False))
plt.title("Total Crashes by State")
plt.xlabel("Total Crashes")
plt.ylabel("State (Abbreviation)")
plt.yticks(fontsize=5)
plt.show()
```

**Inference for Bar Plot:**

The above graph is Bar plot. This bar plot displays the total number of crashes by state.  SC state has the highest total number of crashes and DC state has the least total number of crashes.

```python
# Scatter Plot
sns.scatterplot(x="alcohol", y="speeding", data=car_crashes, color = "orange")
plt.title("Scatter Plot of Alcohol vs. Speeding")
plt.xlabel("Alcohol-related Crashes")
plt.ylabel("Speeding-related Crashes")
plt.show()
```

**Inference for Scatter Plot:**

The above graph is Scatter plot. This scatter plot shows the relationship between alcohol-related crashes and speeding-related crashes. There appears to be a positive correlation.


```python
# Box Plot
sns.boxplot(x="total", data=car_crashes, color="orange")
plt.title("Box Plot of Total Crashes")
plt.xlabel("Total Crashes")
plt.show()
```

**Inference for Box Plot:**

The above graph is Box plot. This box plot visualizes the distribution of total crashes and identifies potential outliers. The median number of total crashes is around 17.


```python
# Histogram
plt.hist(car_crashes["total"], bins=10, color="orange")
plt.title("Histogram of Total Crashes")
plt.xlabel("Total Crashes")
plt.ylabel("Frequency")
plt.show()
```

**Inference for Histogram:**

The above graph is Histogram. This histogram displays the distribution of total crashes. It appears to be right-skewed, with most states having a lower number of total crashes.


```python
# Line Plot
sns.lineplot(x="total", y="alcohol", data=car_crashes, marker="o", color="orange")
plt.title("Line Plot of Total Crashes vs. Alcohol-Related Crashes")
plt.xlabel("Total Crashes")
plt.ylabel("Alcohol-Related Crashes")
plt.show()
```

**Inference for Line Plot:**

the above graph is Line plot. This line plot shows the relationship between the total number of crashes and the number of alcohol-related crashes. As the total number of crashes increases, the number of alcohol-related crashes also tends to increase. There is a positive association between these two variables.
